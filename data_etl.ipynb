```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store CSV into DataFrame

```python
infection_incidence_df = pd.concat(pd.read_excel('incidence_all counteries_all_years.xls', sheet_name=None), ignore_index=True, sort = True)
infection_incidence_df
```

```python
new_infection_incidence_df = infection_incidence_df[['ISO_code', 'Cname','Disease', '2018','2017','2016','2015','2014','2013','2012','2011']].copy()
Global_infection_incidence_df= new_infection_incidence_df.rename(columns = {"ISO_code": "Country_code", "Cname": "Country_name"})
Global_infection_incidence_df.set_index("Country_code")
infection_incidence= Global_infection_incidence_df.dropna()
infection_incidence.head()
```

```python
Immunization_df = pd.concat(pd.read_excel('coverage_estimates_series.xls', sheet_name=None), ignore_index=True, sort = True)
Immunization_df
```

```python
new_Immunization_df = Immunization_df[['ISO_code', 'Cname','Vaccine', '2018','2017','2016','2015','2014','2013','2012','2011']].copy()
Global_Immunization_df= new_Immunization_df.rename(columns = {"ISO_code": "Country_code", "Cname": "Country_name"})
Global_Immunization_df.set_index("Country_code")
Global_Immunization=Global_Immunization_df.dropna()
Global_Immunization.head()
```

### Connect to local database

```python
rds_connection_string = "@localhost:5432/infection_immunization_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Check for tables

```python
engine.table_names()
```

### Use pandas to load csv converted DataFrame into database

```python
infection_incidence.to_sql(name='infection_incidence', con=engine, if_exists='append', index=False)
```

### Use pandas to load json converted DataFrame into database

```python
Global_Immunization.to_sql(name='global_immunization', con=engine, if_exists='append', index=False)
```
